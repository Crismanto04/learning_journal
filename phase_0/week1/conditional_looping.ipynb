# FUNGSI IF

```python
angka = 10
if angka > 10 :
    print (f'{angka} lebih besar dari 10')
elif angka == 10 :
    print (f'{angka} sama dengan 10')
else : 
    print (f'{angka} lebih kecil dari 10')
```

```python
angka = 10
if angka % 2 == 0 :
    print (f'{angka} adalah angka genap')
else :
    print (f'{angka} adalah angka ganjil')
```

```python
tukang_buah = ['jeruk', 'pisang', 'mangga', 'apel']
if len(tukang_buah) > 3 :
    if 'jeruk' in tukang_buah :
        print ('beli jeruk')
    else :
        print ('mampir aja')
else : 
    print ('skip')
```

```python
nilai = 80
if nilai >= 80 and nilai <= 100 :
    print ("A")
elif nilai >= 65 and nilai < 80 :
    print ("B")
elif nilai >= 40 and nilai < 65 : 
    print ("C")
elif nilai >= 20 and nilai < 40 : 
    print ("D")
elif nilai >= 0 and nilai < 20 : 
    print ("E")
else : 
    print("")
```

```python
nilai = input ("masukan nilai : ")
if nilai.isnumeric() :
    nilai = float(nilai)
    if nilai >= 80 and nilai <= 100 :
        print ("A")
    elif nilai >= 65 and nilai < 80 :
        print ("B")
    elif nilai >= 40 and nilai < 65 : 
        print ("C")
    elif nilai >= 20 and nilai < 40 : 
        print ("D")
    elif nilai >= 0 and nilai < 20 : 
        print ("E")
    else : 
        print("")
else :
    print ("Bukan Nilai")
```

```python
pesanan = []
pesan = ''
while pesan != 'cukup' :
    pesan = input('masukan pesanan anda : ')
    if pesan != 'cukup' :
        pesanan.append(pesan)
```

```python
i = 1
while i <= 10 :
    print (f'7 x {i} = {7*i}')
    i = i +1
```

```python
i = 1
j = 1
while j <= 10 :
    while i <= 10 :
        print (f'{j} x {i} = {j*i}')
        i = i+1
    j = j+1
    i = 1
    print ('*'*20)
```

```python
tukang_buah = ['pepaya', 'mangga', 
'pisang', 'jambu']
i = 0
while i < len(tukang_buah) : 
    print (f'abang ini jual {tukang_buah[i]}')
    i = i+1
```

```python
tukang_buah = ['pepaya', 'mangga', 'pisang', 'jambu']
ket = ['stok dikit', 'stok banyak', 'lagi rame', 'stok dikit']
for buah in tukang_buah:
    print(f'abang ini jual {buah}')
```

```python
for i in range(len(tukang_buah)):
    print (f'abang ini jual {tukang_buah[i]}, ket : {ket[i]}')
```
